```python
# Matplotlib is the most popular plotting library in Python
# Gives you control over every aspect of a figure
# Designed to feel similar to Matlab's graphical plotting
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# allows you to see plots within Jupyter notebook
```

```python
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# functional method
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show() # this removes the output cell marker in red
```

```python
# can also add colors in visuals
plt.plot(x, y, 'r-')
```

```python
# adding labels
plt.plot(x, y)
plt.xlabel('This is the x-axis')
plt.ylabel('This is the y-axis')
plt.title('This is the title of the graph')
```

```python
# multi plots on same graph
plt.subplot(1, 2, 1) # number of rows, number of columns, plot number reference
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# object oriented method

# need to create a blank figure, then add axes dimensions, then add data to visualize

fig = plt.figure()
# blank figure object created

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# now we have blank axes

axes.plot(x, y)

# add labels below
axes.set_xlabel('X LABEL')
axes.set_ylabel('Y LABEL')
axes.set_title('TITLE')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.2, 0.6, 0.6]) # make the axis fit within the other to see how it works
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # back to first version

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title("LARGE PLOT")
axes2.set_title("SMALL PLOT")
```

```python
# creating subplots using object oriented method

fig,axes = plt.subplots()

axes.plot(x, y)
```

```python
# subplots allow you specify number of rows and columns for the figure
# both fig and axes specified because subplots unpack them like a tuple

fig,axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3) # axes are automatically generated based on rows and cols
plt.tight_layout() # fixes spacing between axes in the figure
```

```python
# since axes is just an array of Matplotlib values, you can loop through them to add values to graphs

fig,axes = plt.subplots(nrows = 1, ncols = 2)

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
# we can also index the values in axes to enter specific values

fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

```python
# figure size, aspect ratio, and DPI

# figsize arguments takes in a tuple with the width and height of the figure
# dpi is dots per inch, aka pixel size (one value)

fig = plt.figure(figsize = (3, 2), dpi = 100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
# wider version (8x2 instead of 3x2)

fig = plt.figure(figsize = (8, 2), dpi = 100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
# we can do the same thing with subplots

fig,axes = plt.subplots(figsize = (8, 2))

axes.plot(x, y)
```

```python
fig,axes = plt.subplots(figsize = (8, 2), nrows = 2, ncols = 1)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
# you can save Matplotlib figures in various formats (JPEGs, PDFs, etc.)

fig.savefig('my_figure.png', dpi = 200) # specify format in file name
```

```python
# multiple values and adding legends

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# plot two lines - add labels that can be used by ax.legend function
ax.plot(x, x**2, label = 'X SQUARED')
ax.plot(x, x**3, label = 'X CUBED')

# adding a legend - location allows you to choose where it appears on the graph
ax.legend(loc = 0) 

# loc = 0 allows matplotlib to choose where it thinks the legend should be (values are 0-10)
# can also pass in a tuple to enter specific coordinates for location

# regular labels from before
ax.set_title("TITLE")
ax.set_ylabel("Y LABEL")
ax.set_xlabel("X LABEL")
```

```python
# setting colors

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange')
```

```python
# can also using RGB Hex codes

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = '#FF8C00')
```

```python
# setting line width

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', linewidth = 2) # 1 is default - this is double thickness
```

```python
# lw argument also works

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', lw = 2) # 1 is default - this is double thickness
```

```python
# setting alpha (transparency)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', linewidth = 2, alpha = 0.5)
```

```python
# setting line style

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', linewidth = 2, linestyle = '--') # check online and in notebook for other options
```

```python
# ls argument also works here

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', lw = 2, ls = '--') # check online and in notebook for other options
```

```python
# markers are typically used when there isn't much data available to show actual data points

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', lw = 2, ls = '--', marker = 'o', markersize = 10) # other options available as well
```

```python
# more marker options for customization

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', lw = 2, ls = '--', 
        marker = 'o', markersize = 10, markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green') 
```

```python
# examples from the bootcamp notebook on Matplotlib - use as a reference

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
# axis appearance and plot range

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'orange', lw = 2, ls = '--')

ax.set_xlim([0, 2]) # only showing between x-values 0 and 2 of graph
ax.set_ylim([0, 10]) # only showing between y-values 0 and 10
```

```python
# histograms

from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
# boxplots

data = [np.random.normal(0, std, 100) for std in range(0, 4)]

plt.boxplot(data, vert = True, patch_artist = True)

plt.show() # removes error warnings
```
